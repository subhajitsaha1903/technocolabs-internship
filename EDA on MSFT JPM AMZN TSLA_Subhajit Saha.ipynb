## Train Test Split

```python
print(len(df_AMZN_new))
print(len(df_JPM_new))
print(len(df_MSFT_new))
print(len(df_TSLA_new))
```

```python
train_size_amzn = len(df_AMZN_new) - math.ceil(len(df_AMZN_new) * 0.2)
train_size_msft = len(df_MSFT_new) - math.ceil(len(df_MSFT_new) * 0.2)
train_size_jpm = len(df_JPM_new) - math.ceil(len(df_JPM_new) * 0.2)
train_size_tsla = len(df_TSLA_new) - math.ceil(len(df_TSLA_new) * 0.2)
```

```python
test_size_amzn = len(df_AMZN_new) - train_size_amzn
test_size_msft = len(df_MSFT_new) - train_size_msft
test_size_jpm = len(df_JPM_new) - train_size_jpm
test_size_tsla = len(df_TSLA_new) - train_size_tsla
```

```python
train_size_amzn, test_size_amzn
```

```python
test_ind_amzn = len(df_AMZN_new)- test_size_amzn
test_ind_msft = len(df_MSFT_new) - test_size_msft
test_ind_jpm = len(df_JPM_new) - test_size_jpm
test_ind_tsla = len(df_TSLA_new) - test_size_tsla
```

```python
train_amzn = df_AMZN_new.iloc[:test_ind_amzn]
test_amzn = df_AMZN_new.iloc[test_ind_amzn:]
```

```python
train_msft = df_MSFT_new.iloc[:test_ind_msft]
test_msft = df_MSFT_new.iloc[test_ind_msft:]
```

```python
train_jpm = df_JPM_new.iloc[:test_ind_jpm]
test_jpm = df_JPM_new.iloc[test_ind_jpm:]
```

```python
train_tsla = df_TSLA_new.iloc[:test_ind_tsla]
test_tsla = df_TSLA_new.iloc[test_ind_tsla:]
```
