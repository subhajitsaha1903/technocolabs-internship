```python
import pandas as pd
import numpy as np
import os
```

```python
# Set up TWINT config
c = twint.Config()

# c.Search = "Oneplus 9 pro"
c.Username = "Google"
c.Lang = "en"
# config.Limit = 100
# c.Since = '2021–04–01 00:00:00'
# c.Until = '2021-09–25 00:00:00'
c.Since = input('Date (format: "%Y-%m-%d %H:%M:%S"): ')


# Custom output format
c.Limit = 10000
c.Pandas = True

twint.run.Search(c)
```

```python
def column_names():
    return twint.output.panda.Tweets_df.columns
def twint_to_pd(columns):
    return twint.output.panda.Tweets_df[columns]
```

```python
column_names()
```

```python
tweet_df = twint_to_pd(["date", "username", "tweet"])
# tweet_df = twint_to_pd(["date", "tweet"])
```
